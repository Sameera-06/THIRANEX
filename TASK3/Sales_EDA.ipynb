```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

print("Libraries Loaded Successfully")
```

```python
df = pd.read_csv("Superstore.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(10,6))

sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Matrix")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Category',
    y='Sales',
    data=df
)

plt.title("Sales by Category")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    x='Category',
    y='Profit',
    data=df
)

plt.title("Profit by Category")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Sales'],
    bins=30,
    kde=True
)

plt.title("Sales Distribution")

plt.show()
```

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(
    ascending=False
).head(10)

plt.figure(figsize=(12,5))

top_states.plot(kind='bar')

plt.title("Top 10 States by Sales")

plt.ylabel("Sales")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='Segment'
)

plt.title("Customer Segments")

plt.show()
```
